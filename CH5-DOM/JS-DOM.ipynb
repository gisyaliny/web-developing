# JavaScript DOM
The role-play within a web server


### DOM stands for Document Object Model (Right one)


## Access html elements 
* by tag name


* by class name


* by id


* select by query


### Brainstorm
```
    <ul id="sp-list">
        <li class="list">First</li>
        <li class="list">Second</li>
        <li>Third</li>
    </ul>
```
get the `<li>` with `class = "list"`



## Modify properties of elements
```
document.querySelector('h1').style.color = "red"
document.querySelector('#title').style.background = "blue"
```



### Get a class list
```
<body>

    <h1 class="title">Hello</h1>
    
</body>
```


### Dynamic change properties
**You could also `add/remove` a `class` to it**
* html `<h1 >Hello</h1>`
* css 
```.title {
    background-color: burlywood;
}```   
* JS `document.querySelector('h1').classList.add('title')`


you could also use method `toggle()`
```
document.querySelector('button').addEventListener("click",function(){
document.querySelector('h1').classList.toggle('title');
})
```

Once you click this button, it would automatically remove or add a class to `<h1>`, if you use this class in `.css` file, the style from `.css` would be automatically applied

### Dynamic change the content of elements

* Use `innerHTML`    
* If you only want to modify text, could use `innerText`
```
document.querySelector('h1').innerHTML = 'Goodbye';
```

## Using variables to save the selector
```
var title = document.querySelector('#title');
title
```


## Get attributes of elements
* get all attributes, using `document.querySelector('ul').attributes`
* get specific attribute, using `document.querySelector('ul').getAttribute('id')`

### Change the value of attributes
* `document.querySelector('li').attributes.class.value = 'title';`   
It would automatically match the theme for the `title` class from the `.css` file


* `document.querySelector('h1').setAttribute('class','title');`   
Same to the above one, change the value of `class` to `title`

```
var name = 'DasAuto';
console.log(name);
name = 'Simon';
console.log(name)
let a = 1;
console.log(a);
```



### constant variable in JavaScript
* once give a value to it, can not be changed anymore.

```
const Pi = 3.14;
console.log(Pi);
Pi = 3.14159;
console.log(Pi)
```



### Scope of variables in JavaScript
* `var` creates global variables
* `let` and `const` creates local variables

```
var a = 19;

if (true) {
    var b = 20;
}

if (true) {
    let c = 21;
}

do {
    a++;
    console.log(a)
} while (a<= b);

do {
    b++;
    console.log(b)
} while (b<= c);
```



## 5.Operation in JavaScript
#### 5.1 Arithmetic operation in JavaScript
```
var x = 5;

console.log(x);

x += 5;
console.log(x);

x++;
console.log(x);

x--;
console.log(x);
```



#### 5.2 In JavaScript, using `+` to connect one string with one numeric variable are allowed.
```
var x;

x = '3' + 4 + 5; // x = '3' + '4' + '5'
console.log(x);

x = 3 + 4 + '5'; // x = 3 + 4 then x = '7' + '5' 
console.log(x);
```


####  5.3 `==` and `===` in JavaScript
* `==` would do the type transformation automatically
* `===` would do not do the type transformation

```
console.log(123 == '123'); // '123' '123'
console.log(123 === '123');
```


## 6.Flow control in JavaScript

### 6.1 `if`
```
var name = 'Daniel';
var age = 20;
if (name == 'Simon' && age == 21) {
    console.log('Simon');
} else  if (name == 'John'){
    console.log('John');
} else {
    console.log('Stranger!');
}
```


### 6.2 `while` and `do while` loop
* `do while` loop would be excuted at least one time
* both rely on the condition status

```
var a = 19;
var b = 20;
var i = 0;
while ((a +i) <= b) {
    console.log('true')
    i++;
}

do {
    a++;
    console.log(a)
} while (a<= b);
```



### 6.3 `for` loop
* similar to Java

```
for(let i = 0; i<=5; i++) {
    console.log(i)
}
```



### 6.4 Ternary operation
```
var age = 20;
var allowed = age >=18 ? 'Yes' :'NO';
console.log(allowed);
```


### 6.5 `Switch` in JavaScript
```
var name = 'Daniel';

switch(name) {
    case 'Simon':
        console.log("It's Simon");
        break;
    case 'John':
        console.log("It's John");
        break;
    default:
        console.log("It's a stranger");
}
```



## 7. `object` in JavaScript
* using `new Object()` to create
* using `{}` to create
* json

```
var obj = new Object();
var obj2 = {};
```




## 8. `Arrary` in JavaScript
* similar to list in Python
* need to initialize
* could contain `empty`

```
var a = new Array();
var b = [];
a[0] = 'dog';
a[1] = 'cat';
a[5] = 'tiger';
console.log(a);
console.log(a[2]);
```



### Traverse array
```
b = ['dog', 'cat', 'tiger'];

for (let i = 0; i < b.length; i++){
    console.log(b[i]);
}
```


#### we could also use `for in` loop, but would ignore the `empty`
```
for(let i in a){
    console.log(a[i]);
}
```



### Function in array
* `push`, same to `append`
* `pop`
* `shift`: delete the first element
* `unshift`: add elements as the first position in array
* `reverse`
